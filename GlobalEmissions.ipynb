```python
import pandas as pd

```

```python
# You might need to adjust the path to your dataset or load it from the appropriate source
df = pd.read_csv('Methane_final.csv')

# Group by country and calculate the average emissions
average_emissions_per_country = df.groupby('country')['emissions'].mean().reset_index()

# Display the result
print(average_emissions_per_country)
```

```python
# 1. Bar Chart: Average Methane Emissions per Country
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'average_emissions_per_country' DataFrame from previous code
plt.figure(figsize=(12, 6))
sns.barplot(x='country', y='emissions', data=average_emissions_per_country)
plt.xticks(rotation=90)
plt.title('Average Methane Emissions per Country')
plt.xlabel('Country')
plt.ylabel('Average Emissions (kt)')
plt.show()
```
